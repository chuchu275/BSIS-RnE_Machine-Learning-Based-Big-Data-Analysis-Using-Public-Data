```python
df1=pd.read_csv('/content/drive/MyDrive/전국 범죄건수.csv',encoding='cp949')
df2=pd.read_csv('/content/drive/MyDrive/치안센터_최종본.csv',encoding='cp949')
df3=pd.read_csv('/content/drive/MyDrive/R&E채승현.csv',encoding='cp949')
```

```python
df1.columns=['행정구역', '절도', '살인', '강도', '폭행', '성폭력']
```

```python
df1['총합']=df1['절도']+df1['살인']+df1['강도']+df1['폭행']+df1['성폭력']
```

```python
df3.columns=['행정구역', '총인구수', '세대수', '세대당 인구',
       '총인구수', '연령구간인구수', '0~9세',
       '10~19세', '20~29세', '30~39세',
       '40~49세', '50~59세', '60~69세',
       '70~79세', '80~89세', '90~99세',
       '100세 이상', '기초생활수급자수', '평균 공시지가', '전체인구', '도시인구',
       '비도시인구', '도시지역 인구비율', '면적', '인구밀도']
```

```python
df1=df1[df1['행정구역'].isin(df2['행정구역'])]
df2=df2[df2['행정구역'].isin(df1['행정구역'])]
df3=df3[df3['행정구역'].isin(df1['행정구역'])]
len(df1)
```

```python
df1=df1.reset_index()
df2=df2.reset_index()
df3=df3.reset_index()
df1=df1[['절도', '살인', '강도', '폭행', '성폭력', '총합']]
df2=df2[['행정구역', '치안센터수']]
df3=df3[['전체인구','면적']]
```

```python
df4=pd.concat([df2,df1,df3],axis=1)
df4
```

```python
for i in ['절도', '살인', '강도', '폭행', '성폭력']:
  df4[i+'율']=df4[i]/df4['전체인구']*100000
df4
```

```python
df4['치안센터비율인구']=df4['치안센터수']/df4['전체인구']*100000
df4['치안센터비율면적']=df4['치안센터수']/df4['면적']*1000000
```

```python
df4['범죄률']=df4['총합']/df4['전체인구']*100000
df4['범죄률']
```

```python
df4['범죄률'].corr(df4['치안센터수'])
```
